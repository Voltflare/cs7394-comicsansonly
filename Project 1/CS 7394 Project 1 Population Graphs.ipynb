```python
from matplotlib import pyplot as plt
import numpy as np
import math
%matplotlib inline
```

```python
#Simple Exponential- Initial Population * E^(0.01*t)
x = np.arange(0,101,5)
y = 7953952567 * np.power(math.e,(0.01*x))
#graph needs to be 2022-2122 instead of 0-100
x2 = x + 2022
plt.plot(x2,y)
plt.title("Population Growth (Simple Exponential, r = 1%)")
plt.xlabel("Year")
plt.ylabel("Population (in 10 billions)")
plt.show()
```

```python
#Carrying Capacity- CC / (1 + (CC - Initial Population) / Initial Population) * E^(-0.01 * t)

x3 = np.arange(0,101,5)
y2 = 10500000000/ ( (1 + (10500000000 - 7953952567)* np.power(math.e,(-0.01*x3))/7953952567))
#graph needs to be 2022-2122 instead of 0-100
x4 = x3 + 2022
plt.plot(x4,y2)
plt.title("Population Growth (Logistic, K = 10.5 billion)")
plt.xlabel("Year")
plt.ylabel("Population (in billions)")
plt.show()
```
